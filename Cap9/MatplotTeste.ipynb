```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series, DataFrame

plt.figure(1,figsize=(20,5))
a=pd.read_html('https://www.formula1.com/en/results.html/2021/team.html') 
championship=a[0].loc[:,['Team','PTS']].set_index('Team').sort_values(by='PTS')
teams=[team for team in championship.index]
points=[points for points in championship.loc[:,'PTS']]
barra=plt.barh(teams, points)
l=['aquamarine','darkblue','orange','r','b','grey','darkgreen','deepskyblue','crimson','r']
l.reverse()
for i in range(10):
    barra[i].set_fc(l[i])  #Forma mais concisa de se definir individualmente as cores de barras
plt.title('Points Constructors Championship')
for i, v in enumerate(points):
    plt.text(v-11,i-0.3,v, color='white', fontname='Times New Roman',fontsize=20)
```

```python
l.reverse()
barra=plt.pie(points, labels=teams, colors=['aquamarine','darkblue','orange','r','b','grey','darkgreen','deepskyblue','crimson','r'].reverse())
plt.annotate()
```
