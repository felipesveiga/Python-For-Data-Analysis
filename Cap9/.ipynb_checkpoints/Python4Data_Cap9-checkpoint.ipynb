```python
plt.style.use('fivethirtyeight')
laboratorio=pd.read_csv('laboratorio.csv',skipinitialspace=True)
laboratorio.set_index(np.arange(7), inplace=True)
laboratorio.index.name='Hora'
laboratorio
plt.plot(np.arange(7),laboratorio['Bactéria A'], color='darkblue', label='Bactéria A', linewidth=2.5)
plt.plot(np.arange(7), laboratorio['Bactéria B'], color='darkred',label='Bactéria B', linewidth=2.5)
plt.ylabel('População da Cultura')
plt.xlabel('Hora')

populacao_alvo=500 #Podemos utilizar um valor alvo do eixo y para ser destacado pelo matplot


plt.fill_between(np.arange(7),laboratorio['Bactéria A'], populacao_alvo,
                where=(laboratorio['Bactéria A']>populacao_alvo),
                interpolate=True)

'Aqui, destacamos de maneira mais exclusiva o momento em que as Bactérias A superam a marca de 500 indivíduos
                                                                
plt.legend(loc=(0.03,0.8))
```

```python
plt.figure(1,figsize=(12,4))
gm=np.array([200,300,110,50,9,3.5,75,93,125,150])
gm_acoes=pd.DataFrame(gm, index=np.arange(2006,2006+len(gm)), columns={'Preço'})
preco=gm_acoes['Preço']
plt.plot(np.arange(2006,2016),preco, label='Preco GM')
plt.fill_between(np.arange(2006,2016),preco,100, where=(preco<100), 
                 interpolate=True,color='orange',alpha=0.7,label='Crise de 2008')
plt.fill_between(np.arange(2013,2016),preco.loc[2013:2016], preco.loc[2015], where=(preco.loc[2015]>preco.loc[2013:2016]), 
                 interpolate=True,color='red',alpha=0.7,label='Início Recuperação')

plt.title('Ações GM')
plt.xlabel('Ano')
plt.ylabel('$ Billion')
plt.legend(loc='upper right')
```

```python
#Outras funcionalidades do scatter plot
plt.figure(10,figsize=(10,5))

pop=[21,31,12,31,45,2]
area=[700,800,550,1000,890,780]
s=[(a+b)/2 for a,b in zip(pop,area)]
cidades=['Belo Horizonte','Rio de Janeiro','Manaus','Vitória','São Paulo','Macapá']

plt.scatter(pop,area,s=s, edgecolor='k', linewidth=2.5)

'Aqui, plotaremos um scatter plot cujo tamanho dos ícones será proporcional à densidade populacional das cidades'

plt.axis([0,50,400,1100])
for populacao, km, cidade in zip(pop,area, cidades):
    plt.text(populacao-3, km+50, cidade)

'Aqui, estamos inserindo os nomes das cidades estudadas no gráfico'


plt.xlabel('População')
plt.ylabel('Área')
plt.title('Densidade Populacional')
```

```python
#Lidando com Time Series no Matplot

import pandas as pd
import matplotlib.pyplot as plt
from datetime import timedelta, datetime
from matplotlib import dates
plt.style.use('seaborn-paper')

plt.figure(10,figsize=(10,5))

datas=[
    datetime(2020,5,1),
    datetime(2020,5,2),
    datetime(2020,5,3),
    datetime(2020,5,4),
    datetime(2020,5,5)]
y=[200,130,110,50,52]


plt.plot_date(datas,y, linestyle='solid', marker=None)
plt.xticks(ticks=datas)
plt.xlabel('Date', fontsize=12)
plt.ylabel('$ Billions',fontsize=12)
plt.title('GM Stock Value', fontsize=18)

#Como deixar as datas inclinadas
plt.gcf().autofmt_xdate()

#Vamos formatar as datas apresentadas (Ex: Jan, 5 2021)
date_format=dates.DateFormatter('%b,%d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
```

```python
#Montando um plot_date com um DF

'1°: Transformando a coluna de datas em datetime'
plt.style.use('seaborn')
stock=pd.read_csv('stock.csv', skipinitialspace=True)
stock['Date']=pd.to_datetime(stock['Date'])
stock.sort_values(by='Date', inplace=True)

'2°: Fazendo as alterações estéticas'
plt.plot_date(stock['Date'],stock['Close'], linestyle='solid')
plt.gcf().autofmt_xdate()
date_format=dates.DateFormatter('%b, %d, %Y')
plt.gca().xaxis.set_major_formatter(date_format)

'3°: Últimos ajustes'
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Stock Prices')
```

```python
# Agora, vamos plotar gráficos animados em tempo real
import pandas as pd
import random
from itertools import count
import matplotlib.pyplot as plot
from matplotlib.animation import FuncAnimation

x_vals=[]
y_vals=[]

index=count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0,5))
    plt.cla()
    plt.plot(x_vals, y_vals)

ani=FuncAnimation(plt.gcf(), animate, interval=1000)
```

```python
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()


def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0,5))
    plt.cla()
    plt.plot(x_vals, y_vals)


ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()
#plt.show()
```
