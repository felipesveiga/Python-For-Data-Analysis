```python
pattern=re.compile(r'\w+@\w+\.(com|br|org)')  # + é equivalente a 1 ou mais
with open('clientes_contatos.txt', 'r') as f:
    contents=f.read()
    matches=pattern.finditer(contents)
    for match in matches:
        print(match)
```

```python
pattern=re.compile(r'M(r|s|rs)\.?\s[A-Z](\w* \w*)?') #? indica que o caractere pode ter 0 ou 1 unidade
with open('bilderberg.txt', 'r') as f:   #* indica uma variacao de 0 ao infinito
    contents=f.read()                    # () cria um grupo, que pode conter argumentos opcionais (| significa 'ou')
    matches=pattern.finditer(contents)
    for match in matches:
        print(match)
```

```python
#pattern=re.compile(r'([A-Za-z0-9\_]+)')
pattern=re.compile(r'([A-Z])+([a-z])+([0-9])+([\_])?')
with open('password.txt','r') as f:
    contents=f.read()
    matches=pattern.finditer(contents)
    for match in matches:
        print(match)
```

```python
import re 
pattern=re.compile(r'(\w+|\w{1})\.?\s?(\w+|\w{1})\.?\s?(\w+|\w{1})')
with open ('felipe.txt', 'r') as f:
    contents=f.read()
    matches=pattern.finditer(contents)
    for match in matches:
        print(match)
```

```python
pattern=re.compile(r'(([A-Z][a-z][0-9])+([-_])*)')
with open('password.txt','r') as f:
    contents=f.read()
    matches=pattern.finditer(contents)
    for match in matches:
        print(match)
```
