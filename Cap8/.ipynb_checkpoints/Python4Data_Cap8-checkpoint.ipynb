```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
f1_1=pd.read_csv('preseason.csv', skipinitialspace=True)
f1_2=pd.read_csv('preseason2.csv', skipinitialspace=True)
f1_1
pd.merge(f1_1,f1_2, left_on= 'Piloto', right_on='Piloto').set_index('Piloto') 
```

```python
voo=pd.read_csv('voo.csv', skipinitialspace=True)
voo2=pd.read_csv('voo2.csv', skipinitialspace=True)
aviao=pd.merge(voo,voo2,left_on='Companhia', right_on='Empresa', how='outer')
aviao.set_index(['Companhia', 'Empresa']).swaplevel('Companhia','Empresa')
#aviao.replace(aviao.iloc[1,2],200)
```

```python
f1_1=pd.read_csv('preseason.csv', skipinitialspace=True)
f1_2=pd.read_csv('preseason2.csv', skipinitialspace=True)
pre_season=pd.merge(f1_2, f1_1, on=[ 'Piloto', 'Equipe'], how='outer', sort=True, indicator=True)
pre_season.set_index(['Piloto','Equipe'])#.swaplevel('Equipe','Piloto')
```

```python
sorvete1=pd.read_csv('sorvete1.csv', skipinitialspace=True)
sorvete2=pd.read_csv('sorvete2.csv', skipinitialspace=True)
sorvete1
#sorvete=pd.merge(sorvete1,sorvete2, left_on='Marca', right_on='Companhia', how='outer')
sorvete=pd.merge(sorvete1,sorvete2, on=['Marca','Sabor'], how='outer')
#sorvete.set_index(['Marca', 'Sabor'])
sorvete.set_index(['Marca','Sabor']).sort_index(level=[0,1])
```

```python
sorvete1=pd.read_csv('sorvete1.csv', skipinitialspace=True, index_col=[0,1])
sorvete2=pd.read_csv('sorvete2.csv', skipinitialspace=True, index_col=[0,1])
sorvete=pd.merge(sorvete1,sorvete2, left_index=True, right_index=True) #Avisando o Python: as colunas em comum são o index
sorvete.sort_index(level=[0,1]) #Os levels são ordenados de dentro para fora do DataFrame
```

```python
temperatura=pd.read_csv('temperatura.csv', skipinitialspace=True)
umidade=pd.read_csv('umidade.csv' , skipinitialspace=True, index_col=[0,1])
clima=pd.merge(temperatura, umidade, left_on=['Cidade','Ano'], right_index=True, how='inner')
clima
```

```python
f1_1=pd.read_csv('preseason.csv', skipinitialspace=True, index_col=[0,1])
f1_2=pd.read_csv('preseason2.csv', skipinitialspace=True, index_col=[0,1])
f1_1.join(f1_2, how='outer')  #O método join é bastante útil para aglutinar DataFrames de mesmos indexes
```

```python
sudeste=pd.read_csv('sudeste.csv', skipinitialspace=True, index_col=0)
sul=pd.read_csv('sul.csv',skipinitialspace=True, index_col=0)
centro_oeste=pd.read_csv('centro-oeste.csv',skipinitialspace=True, index_col=0)
clima=sudeste.join([sul,centro_oeste], how='outer') #Use Brackets para aglutinar mais de dois DataFrames

clima.rename(columns={'São Paulo':'Bahia', 'Florianópolis':'Amazonas'})
```

```python
s1=pd.Series([1,2],index=['a','b'])
s2=pd.Series([3,4],index=['c','d'])
s3=pd.Series([5,6], index=['e','f'])
s4=pd.Series([7,8,9,10], index=['a','b','x','y'])

pd.concat([s1,s2,s3], axis=0)  
# Series a serem concatenadas entre brackets
#Quando axis=1, cada Series do DataFrame formado é representada por uma das colunas (s1=0, s2=1, s3=2)

#pd.mege(s1,s2, left_index=True, right_index=True)  #As Series não possuem o método join
# Series raramente possuem nomes em seus eixos, então merge também não é a melhor opção
```

```python
s1=pd.Series([1,2],index=['a','b'])
s2=pd.Series([3,4],index=['c','d'])
s3=pd.Series([5,6], index=['e','f'])
s4=pd.Series([7,8,9,10], index=['a','b','x','y'])
pd.concat([s1,s4], axis=1, join='inner') #Em concat, ao invés de 'how', se utiliza 'join'

pd.concat([s1,s4], axis=1)
```
