```python
s1=pd.Series([1,2],index=['a','b'])
s2=pd.Series([3,4],index=['c','d'])
s3=pd.Series([5,6], index=['e','f'])
s4=pd.Series([7,8,9,10], index=['a','b','x','y'])

pd.concat([s1,s2,s3], axis=0)  
# Series a serem concatenadas entre brackets
#Quando axis=1, cada Series do DataFrame formado é representada por uma das colunas (s1=0, s2=1, s3=2)

#pd.merge(s1,s2, left_index=True, right_index=True)  #As Series não possuem o método join
# Series raramente possuem nomes em seus eixos, então merge também não é a melhor opção
```

```python
s1=pd.Series([1,2],index=['a','b'])
s2=pd.Series([3,4],index=['c','d'])
s3=pd.Series([5,6], index=['e','f'])
s4=pd.Series([7,8,9,10], index=['a','b','x','y'])
result=pd.concat([s1,s3], axis=0, join='outer', keys=['s1','s2']) #Em concat, ao invés de 'how', se utiliza 'join'

result.unstack(level=0)
    #Keys possibilitará a identificação da origem dos dados no DataFrame
#pd.concat([s1,s4], axis=1)
```

```python
intencoes_voto=pd.read_csv('intencoes_voto.csv', skipinitialspace=True)
correcao=pd.read_csv('correcao_votos.csv', skipinitialspace=True)
intencoes_corrigidas=intencoes_voto.combine_first(correcao).set_index('Partido')
um=intencoes_corrigidas.unstack() # Com stack, os labels das colunas tornam-se indexes
                                # Perceba que o dropna está como True no stack (não há intenções no PT no Norte)
dois=um.unstack() #Para reverter o stack, utilize unstack (torna um index uma coluna)
um
```
